```python
import requests
from bs4 import BeautifulSoup

# Web sayfasının URL'si
url = 'https://tr.aliexpress.com/item/1005005988698660.html?spm=a2g0o.categorymp.prodcutlist.51.37a1rvgrrvgrk1&pdp_ext_f=%7B%22sku_id%22%3A%2212000035198564762%22%7D&utparam-url=scene%3Asearch%7Cquery_from%3A'

response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    scripts = soup.find_all('script')
    for script in scripts: 
        if 'window.runParams' in script.text: 
            print(script.text)
            break  
else:
    print("Hata")
```
